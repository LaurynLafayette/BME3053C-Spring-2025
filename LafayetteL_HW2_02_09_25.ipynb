```python
bp_readings = [120, 137, 151, 123, 162, 173, 127, 154]

def normalize_bp_readings(readings):
  max_reading = max(readings)
  min_reading = min(readings)
  normalized_readings = []
  for reading in readings:
    normalized_value = (reading - min_reading) / (max_reading - min_reading)
    normalized_readings.append(normalized_value)
  return normalized_readings

print(normalize_bp_readings(bp_readings))
```
